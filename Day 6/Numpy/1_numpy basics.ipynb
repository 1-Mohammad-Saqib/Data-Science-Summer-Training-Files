```python
# Q1 Create an array
# 10 20 30 40 50
# Print it.
import numpy as np 
```

```python
arr = np.array([10,20,30,40,50])
print(arr)
```

```python
# Q2 Print
# First element
# Last element
# Third element
print(arr[0])
print(arr[-1])
print(arr[2])
```

```python
# Q3
# Print
# 20 30 40
# using slicing.
print(arr[1:4])
```

```python
# Q4
# Print
# ndim
# shape
# size
# dtype
print(arr.ndim)
print(arr.shape)
```

```python
print(arr.size)
```

```python
print(arr.dtype)
```
